```python
import pandas as pd
import smtplib
import email.message

# Passo 1: Importar a base de dados

tabela_vendas = pd.read_excel('/home/Vendas.xlsx') # adicionar caminho completo do arquivo "Vendas.xlsx"

# Passo 2: Calcular o faturamento

tabela_faturamento = tabela_vendas[['Valor Final', 'ID Loja']].groupby('ID Loja').sum()

# Passo 3: Calcular a quantidade vendida
tabela_quantidade = tabela_vendas[['Quantidade', 'ID Loja']].groupby('ID Loja').sum()

# Passo 4: Calcular o ticket médio
ticket_medio = (tabela_faturamento['Valor Final'] / tabela_quantidade['Quantidade']).to_frame()
ticket_medio = ticket_medio.rename(columns={0: 'Ticket Medio'})

# Passo 5: Enviar e-mail da tabela geral

tabela_completa = tabela_faturamento.join(tabela_quantidade).join(ticket_medio)

def enviar_email(nome_da_loja, tabela):  
    corpo_email = f"""
    <p>Olá, prezados</p>
    <p>Segue tabela geral com resumo de todas as lojas</p>
    <p>{tabela.to_html()}</p>
    <p>Atenciosamente, </p>
    <p>Seu nome</p>
    """

    msg = email.message.Message()
    msg['Subject'] = f"Relatorio de vendas - {nome_da_loja}"
    msg['From'] = '' # adicionar o e-mail remetente
    msg['To'] = '' # adicionar e-mail destinatário
    password = '' # adicionar senha gerada para apps do e-mail remetente - explicação no arquivo "README.MD"  
    msg.add_header('Content-Type', 'text/html')
    msg.set_payload(corpo_email )

    s = smtplib.SMTP('smtp.gmail.com: 587')
    s.starttls()
    # Login Credentials for sending the mail
    s.login(msg['From'], password)
    s.sendmail(msg['From'], [msg['To']], msg.as_string().encode('utf-8'))
    print('Email enviado')

enviar_email('Diretoria', tabela_completa)

# Passo 6: Enviar e-mail de cada um das lojas

lista_lojas = tabela_vendas['ID Loja'].unique()

for loja in lista_lojas:
    tabela_loja = tabela_vendas.loc[tabela_vendas['ID Loja'] == loja, ['ID Loja', 'Quantidade', 'Valor Final']]
    tabela_loja = tabela_loja.groupby('ID Loja').sum()
    tabela_loja['Ticket Medio'] = tabela_loja['Valor Final'] / tabela_loja['Quantidade']
    enviar_email(loja, tabela_loja)
```
